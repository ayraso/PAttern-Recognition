```python
import numpy as np
import matplotlib.pyplot as plt

import cv2 
```

**TASK 1**

```python
colorTemplate = cv2.imread('MyDrive/Pattern Recognition/Homework01/color_template.png')
colorTemplate = cv2.cvtColor(colorTemplate, cv2.COLOR_BGR2RGB)
plt.imshow(colorTemplate)
```

```python
width, height, channels = colorTemplate.shape
print(f"Width       : {width}")
print(f"Height      : {height}")
print(f"#of Channels: {channels}")
```

```python
colorTemplate_RGBs = np.reshape(colorTemplate, (height*width, channels))
colorTemplate_RGBs = colorTemplate_RGBs.astype(np.float64) / 255.0
colorTemplate_RGBs
```

```python
transformation_matrix_RGB2YIQ = [[0.299,   0.587,   0.114 ],
                                 [0.5959, -0.2746, -0.3213],
                                 [0.2115, -0.5227, -0.3112]]

tm_RGB2YIQ = np.array(transformation_matrix_RGB2YIQ, dtype = np.float64)
tm_RGB2YIQ
```

```python
TIQs = np.dot(colorTemplate_RGBs, tm_RGB2YIQ)
TIQs
```

```python
transformation_matrix_YIQ2RGB = [[1,  0.956,  0.619 ],
                                 [1, -0.272, -0.647 ],
                                 [1, -1.106,  1.703]]

tm_YIQ2RGB = np.array(transformation_matrix_YIQ2RGB, dtype = np.float64)
tm_YIQ2RGB
```

```python
RGBs = np.dot(TIQs, tm_YIQ2RGB.T)
RGBs = (RGBs * 255.0).astype(np.uint8)
RGBs
```

```python
cat1 = cv2.imread('MyDrive/Pattern Recognition/Homework01/cat1.jpg')
cat1 = cv2.cvtColor(cat1, cv2.COLOR_BGR2RGB)
plt.imshow(cat1)
```

```python
flippedVertically_cat1 = cv2.flip(cat1, 0)
plt.imshow(flippedVertically_cat1)
```

```python
flippedHorizantally_cat1 = cv2.flip(cat1, 1)
plt.imshow(flippedHorizantally_cat1)
```

```python
rotatedLeft90_cat1 = cv2.rotate(cat1, cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(rotatedLeft90_cat1)
```

```python
rotatedRight90_cat1 = cv2.rotate(cat1, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(rotatedRight90_cat1)
```

```python
height_cat1, width_cat1 = cat1.shape[:2]
new_height_cat1 = int(height_cat1 / 2)
new_width_cat1 = int(width_cat1 / 2)
resized_cat1 = cv2.resize(cat1, (new_width_cat1, new_height_cat1), interpolation=cv2.INTER_AREA)
plt.imshow(resized_cat1)
```
